## Create New Data Frame

```python
cluster_continuous_date=pd.date_range(start='2018-04-07',end='2019-04-07')
```

```python
cluster_continuous_date
```

```python
cluster_3 = pd.DataFrame(cluster_continuous_date,columns=['cluster_continuous_date'])
```

```python
cluster_3.head()
```

```python
cluster_3.cluster_continuous_date = cluster_3.cluster_continuous_date.astype('datetime64[ns]')
```

```python
cluster_3.dtypes
```

```python
cluster_3.cluster_continuous_date.dt.dayofweek
```

```python
cluster_3['cluster'] = selected_cluster
cluster_3['is_weekend'] = cluster_3.cluster_continuous_date.dt.dayofweek.isin([5,6])
```

```python
cluster_3.head()
```

```python
cluster_3['is_weekday'] = cluster_3.cluster_continuous_date.dt.dayofweek.isin([0,1,2,3,4])
cluster_3['is_start_of_week'] = cluster_3.cluster_continuous_date.dt.dayofweek.isin([0,1])
cluster_3['is_weednesday'] = cluster_3.cluster_continuous_date.dt.dayofweek.isin([2])
cluster_3.head()
```

```python
holiday = pd.read_csv(f'C:/Users/pipatra.NSN-INTRA/Desktop/Anaconda/Projects/21-Sept/Rapido(Like_uber)/data/Holiday.csv')
```

```python
cluster_3['cluster_continuous_date'] = cluster_3.cluster_continuous_date.dt.date
```

```python
cluster_3['is_holiday'] = cluster_3.cluster_continuous_date.isin(holiday.DATE)
```

```python
cluster_3['booking_class'] = 0
```

```python
cluster_3.head()
```

```python
cluster_3 = cluster_3[~cluster_3.cluster_continuous_date.isin(user.date)]
```

```python
cluster_3.head()
```

```python
user.dtypes
```

```python
cluster_3.dtypes
```

```python
cluster_3.rename(columns={"cluster_continuous_date":"date"},inplace=True)
```

```python
complete_data = pd.merge(user,cluster_3,how='outer')
complete_data.head()
```

```python
complete_data.dtypes
complete_data.date = complete_data.date.astype('datetime64[ns]')
```

```python
complete_data.sort_values(by='date')
```
